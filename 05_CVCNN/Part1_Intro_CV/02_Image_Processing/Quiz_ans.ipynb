```python
path = "totoro.png"
```

```python
import cv2
```

```python
def load_img(path):
  img = cv2.imread(path)[:,:,::-1]
  h, w, _ = img.shape
  center = (w//2-1, h//2-1)
  M = cv2.getRotationMatrix2D(center, 45, 1)
  rotate_img = cv2.warpAffine(img, M, (w*2, h*2), borderMode=cv2.BORDER_REFLECT)
  return rotate_img
```
